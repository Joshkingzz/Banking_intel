```python
import pandas as pd

url = r"C:\Users\HP\Desktop\Joshua's files\My data sets\Bank transaction\Banking_DataSet\Bank_raw_data.xlsx"
df = pd.read_excel(url)


# List of columns to encode
columns_to_encode = [
    'TransactionDate', 'TransactionType', 'ProductCategory',
    'ProductSubcategory', 'BranchCity', 'Channel', 'Currency', 'CustomerSegment',
    'RecommendedOffer']

# Assigning unique identifiers to the dataset
for col in columns_to_encode:
    df[col + 'ID'] = df[col].astype('category').cat.codes

cols = ['TransactionID', 'CustomerID', 'TransactionDateID','TransactionDate',
        'TransactionTypeID','TransactionType','Amount', 'ProductCategoryID',
        'ProductCategory', 'ProductSubcategoryID','ProductSubcategory', 'BranchCityID',
        'BranchCity', 'BranchLat', 'BranchLong', 'ChannelID',
        'Channel', 'CurrencyID','Currency', 'CreditCardFees',
        'InsuranceFees', 'LatePaymentAmount', 'CustomerScore', 'MonthlyIncome',
        'CustomerSegmentID','CustomerSegment', 'RecommendedOfferID', 'RecommendedOffer'
        ]

df = df[cols]

df.to_csv('Bank_data.csv', index= False )
```
